```python
import pandas as pd
```

```python
kaggle = pd.read_csv("kaggle.csv")
kaggle.head()
```

```python
hdi = pd.read_csv("HDI.csv")
hdi.head()
```

```python
merged = pd.merge(kaggle, hdi, left_on=["country", "year"], right_on=["Country", "Year"], how="left").drop(["Country", "Year"], axis=1)
merged
```

```python
kaggle["country"].unique()
```

```python
us = merged[merged["country"] == "United States"]
us
```

```python
jp = merged[merged["country"] == "Japan"]
jp
```

```python
merged.columns
```

```python
missing_us = merged[(merged["country"] == "United States") & (merged["HDI for year"].isnull())]
missing_us["HDI for year"] = missing_us["HDI"]
missing_us.drop("HDI", axis = 1, inplace = True)
missing_us #missing 1986-1989
```

```python
missing_jp = merged[(merged["country"] == "Japan") & (merged["HDI for year"].isnull())]
missing_jp["HDI for year"] = missing_jp["HDI"]
missing_jp.drop("HDI", axis = 1, inplace = True)
missing_jp #missing 1986-1989
```

```python
test_us = us[us["HDI for year"].notnull()]
merged_us = pd.concat([test_us, missing_us]).drop("HDI", axis=1)
merged_us
```

```python
test_jp = jp[jp["HDI for year"].notnull()]
merged_jp = pd.concat([test_jp, missing_jp]).drop("HDI", axis=1)
merged_jp
```

```python
both = pd.concat([merged_us, merged_jp]) #still has some missing HDI values
both
```

```python
both.columns
```

```python
both = both[['country','year','sex','age','suicides_no','population','suicides/100k pop','country-year','HDI for year',' gdp_for_year ($) ','gdp_per_capita ($)','generation']]
both 
```

```python
both["gdp_for_year ($)"] = both[" gdp_for_year ($) "].str.replace(',', '')
both.drop(" gdp_for_year ($) ", axis=1, inplace=True)
both
```

```python
both.to_csv("suicide_rates.csv", index=False)
```
